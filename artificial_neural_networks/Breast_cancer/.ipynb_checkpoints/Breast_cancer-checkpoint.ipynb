```python
import pandas as pd 
import sklearn as sk
from keras.models import Sequential 
from keras.layers import Dense
```

```python
previsores = pd.read_csv("entradas-breast.csv")
classes = pd.read_csv("saidas-breast.csv")
```

```python
previsores_train ,previsores_test ,classe_train,classe_test = sk.model_selection.train_test_split(previsores,classes,test_size = 0.25)
```

```python
classificador = Sequential()
#iniciando a primeira hidden layer
classificador.add(Dense(units= 16,activation= 'relu',kernel_initializer='random_uniform',input_dim = 30))
#relu é a função de ativação mais utilizada nas hidden layers 

#começando a camada de saida(lembrando: ela tbm e fully conected)
classificador.add(Dense(units= 1,activation= 'sigmoid'))
#como o problema e binário entre 0 e 1 a melhor função de ativação e á sigmoid 

#vamos compilar a nossa rede
classificador.compile(optimizer= 'adam',loss='binary_crossentropy',metrics=['binary_accuracy'])
classificador.fit(previsores_train,classe_train,batch_size= 10, epochs=100)


```
